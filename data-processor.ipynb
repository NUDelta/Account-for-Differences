```python
import pandas as pd

pd.set_option('display.max_rows', None)
```

```python
state_cat_df = pd.read_csv('state_cat.csv')
# order state_cat_df by num_cats descending
state_cat_df = state_cat_df.sort_values(by=['num_cats'], ascending=False, ignore_index=True)
print(state_cat_df.columns)
print('num of states: ', len(state_cat_df))
state_cat_df.head(3)
```

```python
# filter out len('state_cat') is 0 
state_cat_df = state_cat_df[state_cat_df['num_cats'] > 10]
print('num of states after filtering: ', len(state_cat_df))
```

```python
state_cat_dict = {}
for state in state_cat_df['state']:
    this_state_cat = state_cat_df[state_cat_df['state'] == state]['state_cat'].values[0]
    this_state_cat = this_state_cat[2:-2].split('\', \'')
    # print(type(this_state_cat))
    # print(len(this_state_cat))
    # print(this_state_cat[0])
    # print(this_state_cat[-1])
    state_cat_dict[state] = this_state_cat

print(state_cat_dict)
    
```

```python
# write state_cat_dict to a excel file
# each key is a excel spreatsheet name in the file
# each value is a sheet in the file
writer = pd.ExcelWriter('state_cat.xlsx', engine='xlsxwriter')
for state in state_cat_dict:
    df = pd.DataFrame(state_cat_dict[state], columns=[state])
    df.to_excel(writer, sheet_name=state, index=False, header=False)
writer.save()

```
